```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from urllib.request import urlopen
from bs4 import BeautifulSoup

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

#### Confirmed Global Cases Data

```python
url_cnfrmd_global_data='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df_cnfrmd_global_data = pd.read_csv(url_cnfrmd_global_data)
df_cnfrmd_global_data.head()
```

```python
df_cnfrmd_global_data_mod = df_cnfrmd_global_data[["Country/Region","3/25/20","3/26/20"]]
df_cnfrmd_global_data_mod["new_cases"] = df_cnfrmd_global_data_mod["3/26/20"] - df_cnfrmd_global_data_mod["3/25/20"]
df_cnfrmd_global_data_mod["new_cases_percent"] = df_cnfrmd_global_data_mod["new_cases"]*100/df_cnfrmd_global_data_mod["3/25/20"]
df_cnfrmd_global_data_mod.head()
```

### Goobal Death Data

```python
url_death_global_data='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
df_death_global_data = pd.read_csv(url_death_global_data)
df_death_global_data.head()
```

```python
df_death_global_data_mod = df_death_global_data[["Country/Region","3/25/20","3/26/20"]]
df_death_global_data_mod["new_deaths"] = df_death_global_data_mod["3/26/20"] - df_death_global_data_mod["3/25/20"]
df_death_global_data_mod["new_deaths_percent"] = df_death_global_data_mod["new_deaths"]*100/df_death_global_data_mod["3/25/20"]
df_death_global_data_mod = df_death_global_data_mod.rename(columns={"3/25/20": "3/25/20_deaths", "3/26/20": "3/26/20_deaths"})
df_death_global_data_mod.head()
```

### World Population

```python
url = "https://www.worldometers.info/world-population/population-by-country/"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')

title = soup.title
print(title)
```

```python
table = soup.find('table', attrs={'id':'example2'})
table_rows = table.find_all('tr')

l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
population = pd.DataFrame(l)
population.head()

ppl_df = population[[1,2]]
ppl_df.columns=["Country","Population"]
ppl_df = ppl_df.drop(ppl_df.index[0])
ppl_df.head()
```

### Mergining Confirmed casesand deaths 

```python
dfinal = df_cnfrmd_global_data_mod.merge(ppl_df, how='inner', left_on='Country/Region', right_on='Country')
dfinal.head()
```

```python
dfinal = dfinal.merge(df_death_global_data_mod, how='inner', left_on='Country/Region', right_on='Country/Region')
dfinal.head()
```

```python
dfinal['Population'] = dfinal['Population'].str.replace(',', '').astype(int)
```

```python
dfinal.describe()
```

### Cases by population percentage

```python
dfinal["cases_ppl_perc"] = dfinal["3/26/20"]*100/dfinal["Population"]
```
